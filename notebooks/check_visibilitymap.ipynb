```python
from astropy.io import fits
import numpy as np

import matplotlib.pyplot as plt
```

```python
file = "../output/orbit/90days_from2030-05-03_visibilitymap.fits"
hdu = fits.open(file)
m = hdu[1].data

m_values = np.array(m.tolist()).reshape(-1)
```

```python
bins = np.linspace(-0.5, np.max(m_values)+1.5, 30)

fig, ax = plt.subplots()
ax.hist(m_values, bins=bins)
ax.set(xlim=(bins[0], bins[-1]), xlabel='Visibility Count', ylabel='Number of Pixels', title='Visibility Map Histogram')
plt.show()
```
